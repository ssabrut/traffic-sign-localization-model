### Data Loading

```python
import deeplake

train_ds = deeplake.load('hub://activeloop/gtsrb-train')
test_ds = deeplake.load('hub://activeloop/gtsrb-test')
```

### Display 5 Random Images

```python
import PIL
from matplotlib import pyplot as plt
import random

col = 5
_, ax = plt.subplots(1, col, figsize=(20,10))

for i in range(col):
    index = random.randint(0, len(train_ds.images))
    label = train_ds.labels[index].numpy()[0]
    image = PIL.Image.fromarray(train_ds.images[index].numpy())
    ax[i].imshow(image)
    ax[i].set_title('Label: ' + str(label))
    ax[i].xaxis.set_visible(False)
    ax[i].yaxis.set_visible(False)
plt.show()
```
